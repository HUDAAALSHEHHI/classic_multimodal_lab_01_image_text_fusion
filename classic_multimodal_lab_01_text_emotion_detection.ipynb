```python
analyzer = pipeline("sentiment-analysis")
```

```python
captions = [
    "A happy family enjoying a picnic in the park.",
    "A lonely man walking under the rain.",
    "A group of students celebrating their graduation.",
    "A city destroyed after a massive storm."
]

for text in captions:
    result = analyzer(text)[0]
    print(f"{text}\n→ Sentiment: {result['label']} | Confidence: {result['score']:.3f}\n")
```
